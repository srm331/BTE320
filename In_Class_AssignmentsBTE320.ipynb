```python
def getValues(n):
  assert isinstance(n, int), "'n' must be an integer" # or you can say assert type(n) == int
  items = []

  for i in range(n):
    items.append(float(input("Enter a number:")))
  return items

def calcAverage(alist):
  assert isinstance(alist, list), "'alist' must be a list"
  try:
    return sum(alist[2:]) /len(alist[2:])
  except:
    raise ValueError("The length of list cannot be < 2")
  except Exception as e:
    print(f"Error: {e}")

def main():
  n = int(input("Enter a number of items to be added to the list: "))
  if n < 2:
    raise ValueError("There must be at least 3 items in the list")
  alist = getValues(n)
  avg = calcAverage(alist)
  print(f"Average weight: {avg} lbs.")
```

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def display(self, formal=False):
        if formal:
            print(f"{self.last}, {self.first}")
        else:
            print(f"{self.first}")

    def __str__(self):
        return "Instance of class Greeting"
name = Greeting("John", "Doe")
print(name)
name.display()
name.display(formal=True)
```

```python
class Student:
    def __init__(self, name, number):
        self.__name = name
        self.number = number
        self.__courses = []

    def get_name(self):
      return self.__name

    def get_courses(self):
      return self.__courses

    def set_name(self, newName):
      self.__name = newName

    def __len__(self):
      return len(self.__courses)

    def enroll(self, NewCourse):
      if NewCourse not in self.__courses:
        self.__courses.append(NewCourse)
      else:
        print(f"Student {self.__name} has already enrolled in {NewCourse}.")


```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = self.x - other.x ** 2
    dy = self.y - other.y ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__(self, L1, L2):
    self.p1 = Point(L1[0], L2[0])
    self.p2 = Point(L1[1], L2[0])
    self.p3 = Point(L1[0], L2[1])
    self.p4 = Point(L1[1], L2[1])
    self.width = self.width() # equivalent to: Point.distance(self.p1, self.p2)
    self.height = self.height()

  def width(self):
    return self.p1.distance(self.p2)

  def height(self):
    return self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height
```

```python
coords_1 = [0,3]
coords_2 = [3,0]
z = Rectangle(coords_1, coords_2)
z.area()
```
