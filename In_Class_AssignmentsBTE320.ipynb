```python
num = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
op = (input("Numerical operation (eg., +, -, etc.): "))

if op == "+":
    res = num + num2
elif op == "-":
    res = num - num2
elif op == "*":
    res = num * num2
elif op == "/":
    res = num / num2
elif op == "//":
    res = num // num2
elif op == "**":
    res = num ** num2
elif op == "%":
    res = num % num2
else:
    print("Invalid operation")

print(f"Result: {res}")

```
