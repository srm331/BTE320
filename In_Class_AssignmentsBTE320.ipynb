```python
sales = float(input("Enter sales value ($): "))

cogs = 0.4 * sales
gross_profit = sales - cogs

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")

taxes = float(input("Taxes ($): "))
Net_Income = float(input("Net Income ($): "))
net_income = gross_profit - taxes

print(f"Net income: {net_income}$")
```

```python
num = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
op = (input("Numerical operation (eg., +, -, etc.): "))

if op == "+":
    res = num + num2
elif op == "-":
    res = num - num2
elif op == "*":
    res = num * num2
elif op == "/":
    res = num / num2
elif op == "//":
    res = num // num2
elif op == "**":
    res = num ** num2
elif op == "%":
    res = num % num2
else:
    print("Invalid operation")

print(f"Result: {res}")

```
