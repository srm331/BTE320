```python
num = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
op = (input("Numerical operation (eg., +, -, etc.): "))

if op == "+":
    res = num + num2
elif op == "-":
    res = num - num2
elif op == "*":
    res = num * num2
elif op == "/":
    res = num / num2
elif op == "//":
    res = num // num2
elif op == "**":
    res = num ** num2
elif op == "%":
    res = num % num2
else:
    print("Invalid operation")

print(f"Result: {res}")

```

```python
import math

ticketPrice = 10
defaultAudience = 20

print("Ad$\tProfit")

for adMoney in range(0, 201, 25):
  revenue = (defaultAudience + 2 * round(math.sqrt(adMoney ** 0.5))) * ticketPrice
  costs = 200 + adMoney
  profit = revenue - costs
  print(f"{float(adMoney)}\t{float(profit)}")
```

```python
import math

ticketPrice = 10
defaultAudience = 20
adMoney = 0
print("Ad$\tProfit")
while adMoney <= 200:
  revenue = (defaultAudience + 2 * round(math.sqrt(adMoney))) * ticketPrice
  costs = 200 + adMoney
  profit = revenue - costs
  print(f"{float(adMoney)}\t{float(profit)}")
  adMoney += 25
```
